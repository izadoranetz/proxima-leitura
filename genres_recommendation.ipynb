Para obter o TF-IDF:

```python
df['combined_features'] = df['genres'] + ' ' + df['collection'].fillna('') + ' ' + df['author']

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
print("As n palavras extraídas", tfidf.get_feature_names_out())
```

Resulta na matriz de TF-IDF

```python
print("Matriz esparsa do TF-IDF")
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out(), index=data['title']).sample(23, axis=1).sample(10, axis=0)
```

Calcular a similaridade entre vetores

```python
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim, index=df['title'], columns=df['title'])
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim, index=data['title'], columns=data['title'])
print('Shape:', cosine_sim_df.shape)
cosine_sim_df.sample(23, axis=1).round(2)
```

Lógica para encontrar as maiores pontuações

```python
def genre_recomendations(i, B, itens, k=10):
  ix = B.loc[:,i].to_numpy().argpartition(range(-1, -k, -1))

  closest = B.columns[ix[-1:(-k+2):-1]]

  closest = closest.drop(i, errors='ignore')

  return pd.DataFrame(closest, columns=['title']).merge(itens).head(k)
```

```python
print("\nRecomendações para '1984' (incluindo coleção e autor):")
genre_recomendations('1984', cosine_sim_df, df[['title', 'genres', 'author', 'collection']])
```

```python
# Função melhorada para mostrar mais informações
def detailed_recommendations(i, B, itens, k=10):
    ix = B.loc[:,i].to_numpy().argpartition(range(-1, -k, -1))
    closest = B.columns[ix[-1:(-k+2):-1]]
    closest = closest.drop(i, errors='ignore')

    result = pd.DataFrame(closest, columns=['title']).merge(itens).head(k)
    # Adicionar a similaridade
    similarities = [B.loc[i, title] for title in result['title']]
    result['similarity'] = similarities
    return result.sort_values('similarity', ascending=False)

# Testar a função detalhada
print("\nRecomendações detalhadas para 'The Hobbit':")
detailed_recommendations('The Hobbit', cosine_sim_df, df[['title', 'genres', 'author', 'collection']])
```
